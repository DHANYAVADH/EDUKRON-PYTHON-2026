```python
# Question 11: Find the most frequent character in "mississippi"

print("\nQuestion 11: Find the most frequent character in 'mississippi'")

text = "mississippi"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

most_frequent = max(frequency, key=frequency.get)

print("Most frequent character:", most_frequent)
print("Frequency:", frequency[most_frequent])
```

```python
# Question 23: Convert "hello world" to "hElLo WoRlD" (alternating case)


text = "hello world"
result = ""
index = 0

for char in text:
    if char == " ":
        result += " "
    else:
        if index % 2 == 0:
            result += char.lower()
        else:
            result += char.upper()
        index += 1

print(result)
```

```python
# Question 41: Check if string contains balanced parentheses: "((()))"


text = "((()))"

count = 0
balanced = True

for char in text:
    if char == "(":
        count += 1
    elif char == ")":
        count -= 1
    
    if count < 0:
        balanced = False
        break

if count == 0 and balanced:
    print("Balanced parentheses")
else:
    print("Not balanced")
```

```python
# Question 42: Convert "hello world" to Morse code


text = "hello world"

morse = {
'h': '....', 'e': '.', 'l': '.-..', 'o': '---',
'w': '.--', 'r': '.-.', 'd': '-..'
}

result = ""

for char in text:
    if char == " ":
        result += " / "
    else:
        result += morse[char] + " "

print("Morse Code:", result)
```

```python

# Question 43: Find the longest common substring between "programming" and "grammar"


str1 = "programming"
str2 = "grammar"

longest = ""

for i in range(len(str1)):
    for j in range(i + 1, len(str1) + 1):
        substring = str1[i:j]
        if substring in str2 and len(substring) > len(longest):
            longest = substring

print("Longest common substring:", longest)
```

```python
# Question 47: Check if string is a valid IPv4 address: "192.168.1.1"


ip = "192.168.1.1"
parts = ip.split(".")

valid = True

if len(parts) != 4:
    valid = False
else:
    for part in parts:
        if not part.isdigit() or int(part) < 0 or int(part) > 255:
            valid = False
            break

if valid:
    print("Valid IPv4 address")
else:
    print("Invalid IPv4 address")
```

```python
# Question 49: Convert "hello world" to ROT13 encoding


text = "hello world"
result = ""

for char in text:
    if char.isalpha():
        if char.islower():
            result += chr((ord(char) - ord('a') + 13) % 26 + ord('a'))
        else:
            result += chr((ord(char) - ord('A') + 13) % 26 + ord('A'))
    else:
        result += char

print("ROT13:", result)
```

```python
# Question 50: Check if string is a valid credit card number: "4532015112830366"


card = "4532015112830366"

total = 0
reverse_digits = card[::-1]

for i in range(len(reverse_digits)):
    digit = int(reverse_digits[i])

    if i % 2 == 1:
        digit = digit * 2
        if digit > 9:
            digit -= 9

    total += digit

if total % 10 == 0:
    print("Valid credit card number")
else:
    print("Invalid credit card number")
```
